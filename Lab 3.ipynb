This lab focuses on calculating the frequency of occurance for each unqiue word in Shakespeare's "Romeo and Juliet". A dictionary was made that contained each unqiue word and the number of times it occurred in the text. A bar graph of the frequency of each word appearing was drawn to showcase the most 10 frequent words and the least 10 frequent words that appear.

```python
# Wrappers for supported file formats
import string
import matplotlib.pyplot as plt
import math
import numpy as np
```

***
The code below reads the file "romeoandjuliet.txt" and creates a dictionary. The dictionary's keys are of each unqiue word that appears, and the value of each key is the number of occurances.

```python
# Creates a variable to store the opened file to "read_txt"
read_txt = open("romeoandjuliet.txt", "r")
```

```python
# PROCESSING: Creates dictionary of words in text. Within the for loop,
#             it lowercases all the words, removes the punctuation in the
#             text, and splits the string by "spaces"
dictionary = {}
for line in read_txt:
    line = line.strip().lower()
    line = line.translate(line.maketrans("", "", string.punctuation))
    lines = line.split(" ")
    for word in lines:
        if word in dictionary:
            dictionary[word] = dictionary[word]+1
        else:
            dictionary[word] = 1
```

***
The code below deletes the key "white space" and it's value(s) from the dictionary.

```python
del dictionary[""]
```

***
The code below displays the dictionary to the terminal screen.

```python
print ("Word & Occurance\n")
for k,v in dictionary.items():
    print(k," ", v)
```

***
The following section of code will display a **bar graph** that displays the frequency (the word count) of each unqiue word in the text.

```python
x_coor = dictionary.keys()
y_coor = dictionary.values()
plt.rcParams['figure.figsize'] = [10, 20]
plt.ylabel("Frequency (Word Count)")
plt.xlabel("Word")
plt.title('Frequency that Unqiue Words Appear', fontsize = 20)
plt.bar(x_coor, y_coor)
plt.show()
```

The **least frequent** words are those in which are only written **once** in the text. In comparison, the most frequent words are **pronouns and prepositions**, both of which are commonly used in the English language to effectively communicate. The least frequent words are **proper nouns, explicit adjectives or adverbs, and other uncommon nouns and verbs**.

## **Bonus Question**

The follow sections of code below sort the dictionary according to **decreasing word count**.

```python
# PROCESSING: Copies the original dictionary, sorts the values from
#             smallest to largest, then reverses it to largest to smallest

new_dictionary = [(v, k) for k, v in dictionary.items()]
new_dictionary.sort()
new_dictionary.reverse()
new_dictionary = dict([(k, v) for v, k in new_dictionary])

# OUTPUT: Prints the new dictionary (according to decreasing word count)
print ("According to decreasing word count\n\nWord & Occurance\n")
for k,v in new_dictionary.items():
    print(k," ", v)
```

***
The next section of code creates an array of the **rankings of each word** in the new dictionary, an array of the **log(values) of the rankings of each word** in the new dictionary, and an array of the **word count of each ranking** in the new dictionary.

```python
# PROCESSING: Initialize variables
rank = []
rank1 = []
i = 1
j = 1

# PROCESSING: Calculates the rank of the word
for x in range (len(new_dictionary)):
    rank.append(i)
    i = i + 1

# PROCESSING: Calculates the log[rank] of the word
for x in range (len(new_dictionary)):
    rank1.append(math.log(j))
    j = j + 1

# PROCESSING: Transforms the lists into arrays
rank = np.array(rank)
rank1 = np.array(rank1)
word_count = np.array(list(new_dictionary.values()))
```

***
This next section of code plots the **word count vs the rank of the word**. Where the most frequent word has the rank = 1.

```python
x_coor_0 = rank
y_coor_0 = word_count
plt.rcParams['figure.figsize'] = [10, 10]
plt.ylabel("Word Count")
plt.xlabel("Rank")
plt.title('Word Count vs Rank of Word', fontsize = 15)
plt.plot(x_coor_0, y_coor_0)
plt.show()
```

The graph above shows that there is a relationship that resembles **exponential decay** between the word count and the ranking of the word.
***
This section of code calculates the **log[word count]** of the keys in the dictionary. It then transforms the data into an array to be used for the plot.

```python
for k,v in new_dictionary.items():
    v = math.log(v)
    new_dictionary[k]= v
word_count1 = np.array(list(new_dictionary.values()))
```

***
This next section of code plots the **word count vs the rank of the word** on a **log-log scale**.

```python
x_coor1 = rank1
y_coor1 = word_count1
plt.rcParams['figure.figsize'] = [10, 10]
plt.ylabel("log[Word Count]")
plt.xlabel("log[Rank]")
plt.title('log[Word Count] vs log[Rank of Word]', fontsize = 15)
plt.plot(x_coor1, y_coor1)
plt.show()
```

***
This graph demonstrates that the frequency of a word appearing in the text is **inversely proportional** to its rank. This is an example of **Zipf's Law**, which is the typical model for how terms are distributed across the English Language for example. According to Zipf's law, the frequency of the word decreases very rapidly with rank, which creates an inversely proportional relationship. <br> <br> This means that words commonly used to communicate such as **"the"** will appear the most often.
